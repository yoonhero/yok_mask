```python
import tqdm

stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

tokenized_data = []
for sentence in tqdm.tqdm(x):
    tokenized_sentence = okt.morphs(sentence, stem=True) 
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] 
    tokenized_data.append(stopwords_removed_sentence)
```

```python
from collections import Counter
temp = [word  for sentence in tokenized_data for word in sentence]

count = Counter(temp)
```

```python
count = sorted(count.items(), key=lambda x: x[1], reverse=True)
```

```python
count = count[::-1]
count
```

```python
to_clean = """. , " ' - ? : ! ;""".split(" ")

min_frequency = 10
cleaned =[item[0] if item[1] >= min_frequency and item[0] not in to_clean and "." not in item[0] else None for item in count]

def isNone(x):
    return x != None
cleaned = list(filter(isNone, cleaned))
```
